## Exp-feedback - Creating a class that encompases what happens in each trial as well as creating and presenting stimuli ##

```python
# --- Trial class definition ---
#creating a class that encompases all what happens in a trial
# A Moment of gratitude for w3 which I used to help myself a bit
# https://www.w3schools.com/python/python_class_init.asp
class trial:
    #creating the built in method to assign values to the objects
    def __init__(self, window, set_size, condition):
        #set window to a placeholder window, will put trialWindow for object
        self.window = window
        #setting up the timer clock to record time for the rt
        self.respClock = core.Clock()
        #setting the set size to a placeholder so it can be used in the class (so each trial it can check the set size from global condition when used outside)
        self.set_size = set_size
        #setting the class condition to a placeholder so it can be used in the class (so when used outside will be using each trial it can check whether the target is present or absent from the gloabl list)
        self.condition = condition
        # Target position and image
        #setting the x position of the t (target) to a random coordinate between -760 and 760 with a 100 step difference
        self.x_pos_t = random.randrange(-760, 760, 100)
        #setting the y position of the t (target) to a random coordinate between -340 and 340 with a 100 step difference
        self.y_pos_t = random.randrange(-340, 340, 100)
        #sets the target to the T png at the randomly choosen positions
        self.target = visual.ImageStim(
            self.window, image='T.png', units='pix',
            pos=[self.x_pos_t, self.y_pos_t], size=50
        )
        # Create distractors
        #set the distrcators property to the method distractor in the same class and sets the set size of Distractors to the property defined above - set_size
        self.distractors = self.Distractors(self.set_size)
        #set the property keys to an empty list to fill later with the keys pressed
        self.keys = []
        #set initial rt to 0 - just setting up the property at this time to be used
        self.rt = 0
        #set initial accuracy to 0, to change later - just setting up the property at this time to be used
        self.accuracy = 0

    # Creates a function that is a list of distractors, setting inputs as set) size and as self
    def Distractors(self, set_size):
        #creates an empty list
        distractors = []
        #for the range of set size loop through each time and create a new distractor
        for num in range(set_size):
            #setting the x position of the l (distractor) to a random coordinate between -760 and 760 with a 100 step difference
            x_pos_l = random.randrange(-760, 760, 100)
                    #setting the y position of the l (distractor) to a random coordinate between -340 and 340 with a 100 step difference
            y_pos_l = random.randrange(-340, 340, 100)
            #choosing an orientation to random choice from this list
            orientation = random.choice([0, 90, 180, 270])
            #creating the distractor stimuli image with the random x and y positions and the ori set to the random orientation
            distractor = visual.ImageStim(
                self.window, image='L.png', units='pix',
                pos=[x_pos_l, y_pos_l], size=50, ori=orientation
            )
            #append this newly created distractor to the distractors list
            distractors.append(distractor)
        #return the distractors
        return distractors
```

## My Way creating and presenting stimuli ##

```python
#this sets the positions and orientations of all my stimuli and draws them on the screen
def pos_and_ori(target, distract, samp_size):
    #making a list from number -190 to 190 with step of 25 betweeen each number to prevent letters from being stacked on top of each other
    samplelist = list(range(-180, 180, 25))
    #randomly sampling numbers from the samplelist above up to the setsize of the list for the x-values, the last x-value in the list is ofr the target
    x = random.sample(samplelist, samp_size)
    #doing the same for the y values, the last y-value in the list is for the target
    y = random.sample(samplelist, samp_size)
    #this for loop randomly changes the orientation, and position of my distractors
    #for up to the amount of numbers in my set size - 1
    for n in range(0, samp_size - 1):
        #possible orientations
        orientations = [0, 90, 180, 270]
        #randomly choose an orientation for each distractor
        orin = random.choice(orientations)
        #apply the orientation to the distractor
        distract.ori = orin
        #change the x and y position of the distractors to the number in the index of what set size number we are 
        distract.pos = (x[n], y[n])
        #draw the distractor on the window
        distract.draw()
    #for the last value in the samp_size generate the target in the position of the last x and y values
    for n in range(samp_size - 1, samp_size):
        target.pos = (x[n], y[n])
        #draw the target on teh window
        target.draw()
    #returns the distractor and target to draw them on screen
    return distract, target
```

## My Way of running each experiment - showing stimuli and getting response ##

```python
#definition that states while the timer is under the trial duration take key input from select keys and once key is pressed or trial time expires break the loop
#response usually starts at NONE and same with rt and are defined in this function, as there are no response yet, 
#resp is what the first key pressed is stored as
#rt is the response time that goes with when the key was pressed
def KeyGet(trial_duration = 2.0, rt = None, resp = None):
    #start timer
    startTime = core.getTime() #marks the time of the start of each trial when the stimulus is shown
    #while the timer (which is reset for each trial)minus the start time (so it is never negative) is lower than the trial duration and there is currently no response from the user present stimulsu and wait for response for 2 seconds
    while core.getTime() - startTime < trial_duration and resp is None:
        #this sets what the keys are going to be for the trail which is a for absent, d for there, and escape to leave
        #the timeStamped modifier gets the time the key was pressed
        #d = dere (target present)
        #a = absent (target absent)
        #escape to quit program
        keys = event.getKeys(keyList = ['a','d', 'escape'])
        #if a key is pressed then break loop
        if keys:
            #specifically first set variable key to letter pressed and then set rt to the timeStamped (the second item in a list of lists)
            key = keys[0]
            #response time is current time when key is pressed - the startTime of the trial when the stimuli was shown
            rt = core.getTime() - startTime
            #if key is as set resp to a and break loop
            if key == 'a':
                resp = 'a'
                break
            #if key is d set resp to d and break loop
            elif key == 'd':
                resp = 'd'
                break
            #if key is escape quit program
            elif key == 'escape':
                core.quit()
        core.wait(0.01) #prevent CPU overload
    #if timer exceeds 2 seconds and there is no response set resp to no response, rt to NA, 
    if resp is None:
        resp = 'no_response'
        rt = 999 #will write 999 in the file for RTS with no response
    return resp, rt

#takes the response of resp and rt to provide feedback to the user about thier performance
def Response(resp, rt, targ_there):
    if resp == 'd' and targ_there == 1: #saying target is there when it is there
        corr = 1 #set corr to 1, 1 = correct
        feedback = 'Correct!'  #feedback to correct
        response_time = round(rt, 2) #response time to rt (the timetsamped key press to 2 decimal points
    elif resp == 'a' and targ_there == 1: #saying target is absent when it is there
        corr = 0 #set corr to 0, 0 = wrong
        feedback = 'Incorrect!' #feedback to Incorrect
        response_time = round(rt, 2) #response time to rt to 2 decimal places
    elif resp =='a' and targ_there == 0: #saying target is absent when it is absent
        corr = 1 #set corr to 1
        feedback = 'Correct!' #feedback to Correct
        response_time = round(rt, 2) #rt
    elif resp == 'd' and targ_there == 0: #saying that the target is there when it is not present
        corr = 0 #set corr to 0
        feedback = 'Incorrect' #feedback to incorrect
        response_time = round(rt, 2) #round rt to 2 decimals
    elif resp == 'no_response': #if participant times out of trial without giving a response
        corr = 0 #set corr to 0, since did not answer did not get it correct
        feedback = 'No Response' #feedback to no response
        response_time = 'NA' #response time to rt, which since there is no response is NA
    #returns the corr (0 or 1), feedback (correct, not, no response) and response_time (rt)
    return corr, feedback, response_time
```

```python
# --- Run experiment ---
#tracks what trial we are on starting with 0
trialNum = 0
# creates an empty dictonary for avgerage accuracy for each set size condtion
avgAccuracy = {}
#creates an empty dictonary for the average rt computed for each set size condtion
avgRT = {}

#for each of the set size conditions (so once for each condtion)
for set_size in set_sizes:
    #sets accuracy total to 0
    accuTotal = 0
    #sets rt total to 0
    rtTotal = 0
    #shuffles the condtions (present and absent conditions) for each set size condition
    random.shuffle(conditions)
    #for each item in the conditons (so for each trial)
    for condition in conditions:
        #the current trial Object is created from the class and is set to the the current set_size condition and present/absent condition
        currentTrial = trial(trialWindow, set_size, condition)
        #runs that trial fully (and displays it on screen for user to see)
        #and sets the returned self.rt and self.accuracy to rt and acc to store in the file later
        rt, acc = currentTrial.display()
        #adds this trials recorded accuracy to the running total
        accuTotal += acc
        #adds this trials recorded rt to teh running total
        rtTotal += rt
        #adds 1 to the trial number
        trialNum += 1
        #after each trial is over and everything is recorded stores it in the datafile for each trial..
            #TrailNum = what trial just happened
            #conditon = was it present/absent in this trial
            #set_size = what set size conditon this trial was
            #rt = the response time in this trial
            #acc = the trial accuracy (1 = correct, 0 = incorrect)
        dataFile.write('%i,%s,%i,%f,%f\n' % (trialNum, condition, set_size, rt, acc))
    #computes average percentage accuracy for each set size condition from accuTotal determined above divided by number of trials 
    #its times 100 to get it to a percentage
    #store it in the empty dictonary made above with the key being the current set size codntion and the value being average accuracy for condtion
    avgAccuracy[set_size] = (accuTotal / trials) * 100
    #computes average rt for each set size condtion by getting teh rt total divided by total number of trials
    #store it in the emoty rt dictonary made above with the key being the current set size conditon and the value being teh average rt for that condtion
    avgRT[set_size] = rtTotal / trials

#close datafile after all trials are complete
dataFile.close()

# --- Display results ---
#for each set size conditon
for set_size in set_sizes:
    #make a string with the value of average accuracy from the key of the set size condtion
    end_feedback_acc = "Average accuracy: " + str(avgAccuracy[set_size]) + "%"
    #make a string with average rt for that set size conditon (stored as a key/ value in a dictonary)
    end_feedback_rt = "Average RT: " + str(avgRT[set_size])
    #for ending feedback tell them what set size condtion it was and return feedback and of average accuracy and average rt to the user
    end_feedback = (
        f"For set size {set_size}\n{end_feedback_acc}\n{end_feedback_rt}\n\nPress any key to continue"
    )
    #make the feedback text to the end feedback text created above
    feedback.text = end_feedback
    #draw the feedback text on back screen
    feedback.draw()
    #flip the feedback text for the user to see
    trialWindow.flip()
    #wait for users to press the keys for each one to view their feedback one by one
    event.waitKeys()
```
